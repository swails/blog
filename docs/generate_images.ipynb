```python
import matplotlib.pyplot as plt
import seaborn
import numpy as np
import scipy.stats

seaborn.set_theme()
```

```python
process_x = np.arange(101)
process_y = scipy.stats.norm.pdf(process_x, 50, 10.0)
process_y_less = scipy.stats.norm.pdf(process_x, 20, 10.0)
process_y_more = scipy.stats.norm.pdf(process_x, 80, 10.0)
```

```python
fig = plt.figure(1, figsize=(8, 6))
ax = fig.add_subplot(111)
ax.set_ylabel("Productivity")
ax.set_xlabel("Process")
ax.get_xaxis().grid(visible=False)
ax.get_yaxis().grid(visible=False)
ax.set_xticklabels([])
ax.set_yticklabels([])
ax.plot(process_x[20:80], process_y[20:80], '-')
ax.plot(process_x[:50], process_y_less[:50], '--')
ax.plot(process_x[50:], process_y_more[50:], ':')

fig.savefig("software-development-architecture/productivity-vs-process.svg")
```
